This notebook documents my progress in learning how to implement the NRC emotion lexicon to my song lyrics. I am using the syuzhet package: https://cran.r-project.org/web/packages/syuzhet/vignettes/syuzhet-vignette.html.

```r
options(warn=-1)
suppressWarnings({
    libs <- c("tidyverse","syuzhet")
load.libs <- function(x){
  for(i in x){
    if(!require(i,character.only = TRUE )){
      install.packages(i,dependencies = TRUE )
      library(i,character.only = TRUE )
    }
  }
}
load.libs(libs)
})

data<-read.csv("preprocessed_data (version 2).csv",header=T)
```

```r
nrcPlot<-function(text,albumName){
    tokens<-get_tokens(toString(text))
    nrc<-get_nrc_sentiment(tokens)
    return (barplot(
        sort(colSums(prop.table(nrc[, 1:8]))), 
        horiz=TRUE, 
        cex.names=0.7, 
        las=1, 
        main=albumName, xlab="Percentage"))
}

nrcPlot(data[1,2],"Doo-Wops & Hooligans")
nrcPlot(data[2,2],"Unorthodox Jukebox")
nrcPlot(data[3,2],"XX4K")
```
